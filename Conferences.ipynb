```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Files to load
conf_data_to_load = "Conferences.csv"
team_data_to_load = "MTeamConferences.csv"
```

```python
# Read the city data file and store it in a pandas DataFrame.
conf_data_df = pd.read_csv(conf_data_to_load)
conf_data_df.head(10)
```

```python
# Read the city data file and store it in a pandas DataFrame.
team_data_df = pd.read_csv(team_data_to_load)
team_data_df.head(10)
```

```python
# Combine the data into a single dataset
combined_data_df = pd.merge(conf_data_df, team_data_df, how="left", on=["ConfAbbrev", "ConfAbbrev"])

# Display the DataFrame
combined_data_df.head()
```

```python
current_year_df = combined_data_df[combined_data_df["Season"] == 2022]
current_year_df
```

```python
# Export results to csv
current_year_df.to_csv('conferences_2022_results.csv', index=False)
```
