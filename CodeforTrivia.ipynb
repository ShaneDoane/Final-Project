```python
# Import dependencies
import pandas as pd
```

```python
# Open file
file_to_open = 'detailed_all_game_results.csv'
game_results_df = pd.read_csv(file_to_open)

game_results_df.head()
```

```python
# View columns
game_results_df.columns
```

```python
# Filter for just tournament games
tourney_df = game_results_df[game_results_df['GameType'] == 'Tournament']

tourney_df.head()
```

```python
# Get the differences in scores for each game
tourney_df['Spread'] = tourney_df['WScore'] - tourney_df['LScore']

tourney_df.head()
```

```python
# Get the largest spread
tourney_df.sort_values(by=['Spread'], ascending=False)
```
