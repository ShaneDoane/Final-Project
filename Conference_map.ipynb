```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Files to load
conf_data_to_load = "Conferences.csv"
team_data_to_load = "MTeamConferences.csv"
city_data_to_load = "Cities.csv"
game_data_to_load = "MGameCities.csv"
winning_data_to_load = "MRegularSeasonDetailedResults.csv"
```

```python
# Read the conferences data file and store it in a pandas DataFrame.
conf_data_df = pd.read_csv(conf_data_to_load)
conf_data_df.head(10)
```

```python
# Read the team conferences data file and store it in a pandas DataFrame.
team_data_df = pd.read_csv(team_data_to_load)
team_data_df.head(10)
```

```python
# Read the city data file and store it in a pandas DataFrame.
city_data_df = pd.read_csv(city_data_to_load)
city_data_df.head(10)
```

```python
# Read the game data file and store it in a pandas DataFrame.
game_data_df = pd.read_csv(game_data_to_load)
game_data_df.head(10)
```

```python
# Read the winning data file and store it in a pandas DataFrame.
winning_data_df = pd.read_csv(winning_data_to_load)
winning_data_df.head(10)
```

```python
winning_data_df.columns.values.tolist()
```

```python
winning_loc_df = winning_data_df.drop([
 'DayNum',
 'WScore',
 'LScore',
 'NumOT',
 'WFGM',
 'WFGA',
 'WFGM3',
 'WFGA3',
 'WFTM',
 'WFTA',
 'WOR',
 'WDR',
 'WAst',
 'WTO',
 'WStl',
 'WBlk',
 'WPF',
 'LFGM',
 'LFGA',
 'LFGM3',
 'LFGA3',
 'LFTM',
 'LFTA',
 'LOR',
 'LDR',
 'LAst',
 'LTO',
 'LStl',
 'LBlk',
 'LPF'], axis = 1)
```

```python
game_data_df = game_data_df[game_data_df["Season"] == 2022]
game_data_df
```

```python
game_data_df['WTeamID-LTeamID'] = game_data_df['WTeamID'].map(str) + '-' + game_data_df['LTeamID'].map(str)
game_data_df.head()
```

```python
winning_loc_df = winning_loc_df[winning_loc_df["Season"] == 2022]
winning_loc_df
```

```python
winning_loc_df['WTeamID-LTeamID'] = winning_loc_df['WTeamID'].map(str) + '-' + winning_loc_df['LTeamID'].map(str)
winning_loc_df.head()
```

```python
winning_loc_df = winning_loc_df[winning_loc_df["WLoc"] == 'H']
winning_loc_df
```

```python
duplicate_winning_loc_data_removed = winning_loc_df.drop_duplicates(subset=['WTeamID'])
duplicate_winning_loc_data_removed
```

```python
duplicate_game_data_removed = game_data_df.drop_duplicates(subset=['WTeamID'])
duplicate_game_data_removed
```

```python
team_data_df = team_data_df[team_data_df["Season"] == 2022]
team_data_df
```

```python
# Combine the data into a single dataset
combined_TeamID_df = pd.merge(duplicate_winning_loc_data_removed, duplicate_game_data_removed, how="left", on=["WTeamID-LTeamID", "WTeamID-LTeamID"])

# Display the DataFrame
combined_TeamID_df.head()
```

```python
# Combine the data into a single dataset
combined_conf_data_df = pd.merge(conf_data_df, team_data_df, how="left", on=["ConfAbbrev", "ConfAbbrev"])

# Display the DataFrame
combined_conf_data_df.head()
```

```python
combined_TeamID_df.rename(columns={'WTeamID_x': 'TeamID'}, inplace=True)
combined_TeamID_df
```

```python
# Combine the data into a single dataset
combined_city_data_df = pd.merge(combined_conf_data_df, combined_TeamID_df, how="left", on=["TeamID", "TeamID"])

# Display the DataFrame
combined_city_data_df.head()
```

```python
# Export results to csv
combined_city_data_df.to_csv('combined_city_results.csv', index=False)
```
