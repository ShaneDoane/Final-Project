```python
# Import dependencies
import pandas as pd
```

```python
# Open the file
tourney_csv = "MNCAATourneyDetailedResults.csv"
tourney_df = pd.read_csv(tourney_csv)

tourney_df.head()
```

```python
# Add a column of W for winner
tourney_df['OutcomeW'] = 'W'

tourney_df.head()
```

```python
# Add a column of L for loser
tourney_df['OutcomeL'] = 'L'

tourney_df.head()
```

```python
# Stack winning and losing results using concatenate
results_df = pd.concat([
        tourney_df[['Season','DayNum','WTeamID','WScore','OutcomeW','NumOT','WFGM','WFGA','WFGM3','WFGA3','WFTM','WFTA','WOR','WDR','WAst','WTO','WStl','WBlk','WPF']].rename(columns={
            'WTeamID':'TeamID', 'WScore':'Score','OutcomeW':'Outcome','WFGM':'FGM','WFGA':'FGA','WFGM3':'FGM3', 'WFGA3':'FGA3','WFTM':'FTM','WFTA':'FTA','WOR':'OR','WDR':'DR','WAst':'Ast','WTO':'TO','WStl':'Stl','WBlk':'Blk','WPF':'PF'}),
        tourney_df[['Season','DayNum','LTeamID','LScore','OutcomeL','NumOT','LFGM','LFGA','LFGM3','LFGA3','LFTM','LFTA','LOR','LDR','LAst','LTO','LStl','LBlk','LPF']].rename(columns={
            'LTeamID':'TeamID', 'LScore':'Score','OutcomeL':'Outcome','LFGM':'FGM','LFGA':'FGA','LFGM3':'FGM3','LFGA3':'FGA3','LFTM':'FTM','LFTA':'FTA','LOR':'OR','LDR':'DR','LAst':'Ast','LTO':'TO','LStl':'Stl','LBlk':'Blk','LPF':'PF'})
])

# View df
results_df.head(50)
```

```python
# Import reg season results to make one df of all game records
# Open the file
reg_season_csv = "MRegularSeasonDetailedResults.csv"
reg_season_df = pd.read_csv(reg_season_csv)

reg_season_df.head()
```

```python
# Add a column of W for winner
reg_season_df['OutcomeW'] = 'W'

reg_season_df.head()
```

```python
# Add a column of L for loser
reg_season_df['OutcomeL'] = 'L'

reg_season_df.head()
```

```python
# Stack winning and losing results using concatenate
season_results_df = pd.concat([
        reg_season_df[['Season','DayNum','WTeamID','WScore','OutcomeW','NumOT','WFGM','WFGA','WFGM3','WFGA3','WFTM','WFTA','WOR','WDR','WAst','WTO','WStl','WBlk','WPF']].rename(columns={
            'WTeamID':'TeamID', 'WScore':'Score','OutcomeW':'Outcome','WFGM':'FGM','WFGA':'FGA','WFGM3':'FGM3', 'WFGA3':'FGA3','WFTM':'FTM','WFTA':'FTA','WOR':'OR','WDR':'DR','WAst':'Ast','WTO':'TO','WStl':'Stl','WBlk':'Blk','WPF':'PF'}),
        reg_season_df[['Season','DayNum','LTeamID','LScore','OutcomeL','NumOT','LFGM','LFGA','LFGM3','LFGA3','LFTM','LFTA','LOR','LDR','LAst','LTO','LStl','LBlk','LPF']].rename(columns={
            'LTeamID':'TeamID', 'LScore':'Score','OutcomeL':'Outcome','LFGM':'FGM','LFGA':'FGA','LFGM3':'FGM3','LFGA3':'FGA3','LFTM':'FTM','LFTA':'FTA','LOR':'OR','LDR':'DR','LAst':'Ast','LTO':'TO','LStl':'Stl','LBlk':'Blk','LPF':'PF'})
])

# View df
season_results_df.head(50)
```

```python
# Combine both tournament and reg season results
all_game_results_df = pd.concat([
        season_results_df[['Season', 'DayNum', 'TeamID', 'Score', 'Outcome', 'NumOT', 'FGM', 'FGA',
       'FGM3', 'FGA3', 'FTM', 'FTA', 'OR', 'DR', 'Ast', 'TO', 'Stl', 'Blk',
       'PF']],
        results_df[['Season', 'DayNum', 'TeamID', 'Score', 'Outcome', 'NumOT', 'FGM', 'FGA',
       'FGM3', 'FGA3', 'FTM', 'FTA', 'OR', 'DR', 'Ast', 'TO', 'Stl', 'Blk',
       'PF']]
])
    
# View resulting df
all_game_results_df.head()
```

```python
# Check the length
len(all_game_results_df)
```

```python
# Export to csv
all_game_results_df.to_csv('all_game_results.csv')
```
