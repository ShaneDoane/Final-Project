```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# Open the file
tourney_csv = "MNCAATourneyDetailedResults.csv"
tourney_df = pd.read_csv(tourney_csv)

tourney_df.head()
```

```python
# Add a column to designate tournament game
tourney_df['GameType'] = 'Tournament'

tourney_df.head()
```

```python
# Import reg season results to make one df of all game records
# Open the file
reg_season_csv = "MRegularSeasonDetailedResults.csv"
reg_season_df = pd.read_csv(reg_season_csv)

reg_season_df.head()
```

```python
# Add a column to designate regular season game
reg_season_df['GameType'] = 'RegSeason'

reg_season_df.head()
```

```python
# Get columns for concatenating the tables from the tourney df
tourney_df.columns
```

```python
# Get columns for concatenating the tables from the reg season df
reg_season_df.columns
```

```python
# Combine both tournament and reg season results
results_df = pd.concat([
        tourney_df[['Season', 'DayNum', 'WTeamID', 'WScore', 'LTeamID', 'LScore', 'WLoc',
       'NumOT', 'WFGM', 'WFGA', 'WFGM3', 'WFGA3', 'WFTM', 'WFTA', 'WOR', 'WDR',
       'WAst', 'WTO', 'WStl', 'WBlk', 'WPF', 'LFGM', 'LFGA', 'LFGM3', 'LFGA3',
       'LFTM', 'LFTA', 'LOR', 'LDR', 'LAst', 'LTO', 'LStl', 'LBlk', 'LPF',
       'GameType']],
        reg_season_df[['Season', 'DayNum', 'WTeamID', 'WScore', 'LTeamID', 'LScore', 'WLoc',
       'NumOT', 'WFGM', 'WFGA', 'WFGM3', 'WFGA3', 'WFTM', 'WFTA', 'WOR', 'WDR',
       'WAst', 'WTO', 'WStl', 'WBlk', 'WPF', 'LFGM', 'LFGA', 'LFGM3', 'LFGA3',
       'LFTM', 'LFTA', 'LOR', 'LDR', 'LAst', 'LTO', 'LStl', 'LBlk', 'LPF',
       'GameType']]
])

results_df.head()
```

```python
# Confirm df lengths
print(len(tourney_df))
print(len(reg_season_df))
print(len(results_df))
```

```python
# Import the ranking file
rank_csv = "MMasseyOrdinals.csv"
ranking_df = pd.read_csv(rank_csv)

ranking_df.head()
```

```python
# Narrow down to just the SAG ranks (based on Shane's research that they have the most ranks)
SAG_ranks = ranking_df.loc[ranking_df['SystemName']=='SAG']

SAG_ranks.head()
```

```python
# Add week number calculation for ranking and round up using numpy's ceiling function
SAG_ranks['WeekNum'] = (SAG_ranks['RankingDayNum']/7).apply(np.ceil)

SAG_ranks.head()
```

```python
# Add a Season-Week-Team column for merging with results df
SAG_ranks['Season-Week-Team'] = SAG_ranks['Season'].map(str) + '-' + SAG_ranks['WeekNum'].map(str) + '-' + SAG_ranks['TeamID'].map(str)

SAG_ranks.head()
```

```python
# Add week number to results df so we can eventually pull rank from SAG df
results_df['WeekNum'] = (results_df['DayNum']/7).apply(np.ceil)

results_df.head()
```

```python
# Replace tournament NaN with most recent season rankings
results_df['WeekNum'].values[results_df['WeekNum'].values > 19.0] = 19.0
  
results_df.head()

# print(len(results_df))
```

```python
# Add a WSeason-Week-Team and LSeason_Week-Team column for merging with SAG df
results_df['WSeason-Week-Team'] = results_df['Season'].map(str) + '-' + results_df['WeekNum'].map(str) + '-' + results_df['WTeamID'].map(str)

results_df['LSeason-Week-Team'] = results_df['Season'].map(str) + '-' + results_df['WeekNum'].map(str) + '-' + results_df['LTeamID'].map(str)

results_df.head()

# print(len(results_df))
```

```python
# Merge rankings for winning team into the original results_df
results_df = results_df.merge(SAG_ranks[['Season-Week-Team','OrdinalRank']], how='left', left_on='WSeason-Week-Team', right_on='Season-Week-Team').drop(columns=['Season-Week-Team'])

results_df.head(50)

print(len(results_df))
```

```python
# Rename OrdinalRank column to indicate it is for the winning team
results_df.rename(columns={'OrdinalRank': 'WOrdinalRank'},inplace=True)

results_df.head()
```

```python
# Merge rankings for losing team into the original results_df
results_df = results_df.merge(SAG_ranks[['Season-Week-Team','OrdinalRank']], how='left', left_on='LSeason-Week-Team', right_on='Season-Week-Team')

results_df.head()
```

```python
# Drop unnecessary Season-Week_Team column
results_df.drop('Season-Week-Team',axis=1, inplace=True)

results_df
```

```python
# Rename OrdinalRank column to indicate it is for the losing team
results_df.rename(columns={'OrdinalRank': 'LOrdinalRank'},inplace=True)

results_df.head()
```

```python
# Add column for % field goals made
results_df['WFG%'] = results_df['WFGM']/results_df['WFGA']
results_df['LFG%'] = results_df['LFGM']/results_df['LFGA']

results_df.head()
```

```python
# Add column for % threes made
results_df['WFG3%'] = results_df['WFGM3']/results_df['WFGA3']
results_df['LFG3%'] = results_df['LFGM3']/results_df['LFGA3']

results_df.head()
```

```python
# Add a column for % FT made
results_df['WFT%'] = results_df['WFTM']/results_df['WFTA']
results_df['LFT%'] = results_df['LFTM']/results_df['LFTA']

results_df.head()
```
