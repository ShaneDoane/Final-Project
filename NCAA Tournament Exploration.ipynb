```python
# Import dependencies
import pandas as pd
```

```python
# Open the file
tourney_csv = "MNCAATourneyDetailedResults.csv"
tourney_df = pd.read_csv(tourney_csv)

tourney_df.head()
```

```python
# View summary statistics
tourney_df.describe()
```

```python
# Import MTeams.csv to bring in team info
teams_csv = "MTeams.csv"
teams_df = pd.read_csv(teams_csv)

teams_df.head()
```

```python
# Merge winning team name into df
tourney_team_df = tourney_df.merge(teams_df[['TeamName','TeamID']], how='left', left_on='WTeamID', right_on='TeamID').drop(columns=['TeamID'])

tourney_team_df.head()
```

```python
# Rename winning team name column
tourney_team_df.rename(columns={'TeamName':'WTeamName'}, inplace=True)

tourney_team_df.head()
```

```python
# Merge losing team name into df
tourney_team_df = tourney_team_df.merge(teams_df[['TeamName','TeamID']], how='left', left_on='LTeamID', right_on='TeamID').drop(columns=['TeamID'])

tourney_team_df.head()
```

```python
# Rename losing team name column
tourney_team_df.rename(columns={'TeamName':'LTeamName'}, inplace=True)

tourney_team_df.head()
```

```python
# Get columns names for reordering
list(tourney_team_df)
```

```python
# Reorder Columns
tourney_team_df = tourney_team_df.loc[:,['Season',
 'DayNum',
 'WTeamID',
 'WTeamName',                                       
 'WScore',
 'LTeamID',
 'LTeamName',
 'LScore',
 'WLoc',
 'NumOT',
 'WFGM',
 'WFGA',
 'WFGM3',
 'WFGA3',
 'WFTM',
 'WFTA',
 'WOR',
 'WDR',
 'WAst',
 'WTO',
 'WStl',
 'WBlk',
 'WPF',
 'LFGM',
 'LFGA',
 'LFGM3',
 'LFGA3',
 'LFTM',
 'LFTA',
 'LOR',
 'LDR',
 'LAst',
 'LTO',
 'LStl',
 'LBlk',
 'LPF']]

tourney_team_df.head()
```

```python
# Get the teams with the most wins in the tournament
tourney_team_df[["WTeamID","WTeamName"]].value_counts()
```

```python
# Get the teams with the most losses in the tournament
tourney_team_df[["LTeamID","LTeamName"]].value_counts()
```

```python
# Export to Excel to build visuals in Tableau
tourney_team_df.to_csv('tourney_team_df.csv',index=False)
```
